```python
import pandas as pd
df = pd.read_pickle('sentences.pkl')
df['context'] = df['keyword_type'].str.startswith('context')
df[df['context']]
df.to_csv('sentences_sustainability_keywords.csv')
```

```python
pd.set_option('max_rows', 101)
pd.set_option('min_rows', 100)
```

```python
df['keyword'] = df['keyword'].str.lower()
df = df.set_index(['keyword_type', 'keyword']).groupby(['keyword_type', 'keyword', 'sentiment']).agg({'score':['count', 'sum', 'mean']}).reset_index().pivot(index=['keyword_type', 'keyword'], values=[("score",'count'), ("score",'sum'), ("score",'mean')], columns="sentiment")
```

```python
df['total_count'] = df[[(("score",'count'),'negative'), (("score",'count'),'neutral'), (("score",'count'),'positive')]].sum(1)
df['positive_rate'] = df[(("score",'count'),'positive')] / (df[(("score",'count'),'positive')] + df[(("score",'count'),'negative')])
df[df['total_count'] > 50][df['context']].sort_values('positive_rate', ascending=False).set_index(['keyword_type', 'keyword'])
```
