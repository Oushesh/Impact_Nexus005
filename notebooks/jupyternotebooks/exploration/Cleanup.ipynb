```python
from spacy.matcher import DependencyMatcher, Matcher
verb_object_matcher = DependencyMatcher(nlp.vocab)
# Add match ID "HelloWorld" with no callback and one pattern
patterns = [
[
    # anchor token: founded
    {
        "RIGHT_ID": "verb",
        "RIGHT_ATTRS": {"POS": "VERB"}
    },
    {
        "LEFT_ID": "verb",
        "RIGHT_ID": "object",
        "REL_OP": ".*",
        "RIGHT_ATTRS": {"DEP": "pobj"}
    }
],
[
    # anchor token: founded
    {
        "RIGHT_ID": "verb",
        "RIGHT_ATTRS": {"POS": "VERB"}
    },
    {
        "LEFT_ID": "verb",
        "RIGHT_ID": "object",
        "REL_OP": ".*",
        "RIGHT_ATTRS": {"DEP": "dobj"}
    }
]]
verb_object_matcher.add("ProperSentence", patterns)
```

```python
doc = nlp(text[:10000])
```

```python
sents = list(doc.sents)
```

```python
sentences
```

```python
sents[1]
```

```python
sentences = ''
last_i = 0
for i, sent in enumerate(sents[:40]):
    if not verb_object_matcher(sent):
        continue
    sentence = sent.text
    sentence = re.sub(r'(\S)\s*\n\s*([\S])', r'\1 \2', sentence)
    sentence = re.sub(r'\n\n\n+', r'\n\n', sentence)
    sentence = re.sub('^ +', '', sentence)
    if sentence.lower().startswith("Source:"):
        continue
    if re.match(r'^[\d\W ]+$', sentence):
        continue
    if re.match(r'^(?:\b(?:\w){1,2}\b *)+$', sentence):
        continue
    if re.findall(r'\.{5,}', sentence):
        continue
    if re.findall(r'( \S){5,}', sentence):
        continue
    if re.findall(r'^\s*page\b', sentence, flags=re.IGNORECASE):
        continue
    if re.findall(r'^\s*page\b', sentence, flags=re.IGNORECASE):
        continue
    if sentence.strip():
        if i - last_i > 1:
            sentences += "\n"
        last_i = i
        sentences += sentence + '\n'
sentences = sentences.strip()
```

```python
print(sentences)
```

```python
doc = nlp(sentences)
```
